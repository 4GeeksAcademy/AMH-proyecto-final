## En este scrpt se procesarn los datos de df de diferentes variables

```python
#Libraries
import pandas as pd
```

```python
#Funcion que transformar los trimestres en dato transformable a datatime
def obtener_rango_fechas_trimestre(trimestre):
    year = trimestre[3:7]
    t= trimestre[:2]
    if t == 'T1':
        return pd.date_range(start=f'01/01/{year}', end=f'31/03/{year}')
    elif t == 'T2':
        return pd.date_range(start=f'01/04/{year}', end=f'30/06/{year}')
    elif t == 'T3':
        return pd.date_range(start=f'01/07/{year}', end=f'30/09/{year}')
    elif t == 'T4':
        return pd.date_range(start=f'01/10/{year}', end=f'31/12/{year}')
    else:
        return []  # Retorna una lista vacía en caso de trimestre desconocido


    

#Funcion que transforma semestres
def obtener_rango_fechas_semestre(semestre):
    year = semestre[3:7]
    s= semestre[:2]
    if s == 'S1':
        return pd.date_range(start=f'01/01/{year}', end=f'30/06/{year}')
    elif s == 'S2':
        return pd.date_range(start=f'01/07/{year}', end=f'31/12/{year}')
    else:
        return []  # Retorna una lista vacía en caso de trimestre desconocido
```

```python
#Cargar de los csv de cada estacion como df
df_pib=pd.read_csv('../data/interim/PIB_BCN.csv', delimiter=';', encoding='latin1')
df_pob=pd.read_csv('../data/interim/pobl_BCN.csv', delimiter=';')

df_pib
```

```python
df_pob
```

```python
# Convertir las columnas 'trimestre' y 'concepto' a tipo string
df_pib['trimestre'] = df_pib['trimestre'].astype('string')
df_pib['concepto'] = df_pib['concepto'].astype('string')

# Mostrar información del DataFrame
df_pib.info()
```

```python
#generar nuevo df en funcion de dias del PIB
# Crear listas para almacenar los resultados
fechas = []
conceptos = []
valores = []

# Iterar sobre el DataFrame original para llenar el nuevo DataFrame
for index, row in df_pib.iterrows():
    rango_fechas = obtener_rango_fechas_trimestre(row['trimestre'])
    for fecha in rango_fechas:
        fechas.append(fecha)
        conceptos.append(row['concepto'])
        valores.append(row['valor'])

# Crear un nuevo DataFrame con todas las fechas
df_pib_amplio = pd.DataFrame({
    'fecha': fechas,
    'concepto': conceptos,
    'valor': valores
})

df_pib_amplio
```

```python
# Crear nuevo df en funcion de dias de la poblacion 
# Crear listas para almacenar los resultados
fechas = []
valores = []

# Iterar sobre el DataFrame original para llenar el nuevo DataFrame
for index, row in df_pob.iterrows():
    rango_fechas = obtener_rango_fechas_semestre(row['semestre'])
    for fecha in rango_fechas:
        fechas.append(fecha)
        valores.append(row['valor'])

# Crear un nuevo DataFrame con todas las fechas
df_pob_amplio = pd.DataFrame({
    'fecha': fechas,
    'valor': valores
})

df_pob_amplio
```
