## Predicción de de consumo por horas

```python
#Libraries
import pandas as pd
```

```python
#Cargar de los csv como df
df2019=pd.read_csv('../data/interim/2019_consum_electricitat_bcn.csv')
df2020=pd.read_csv('../data/interim/2020_consum_electricitat_bcn.csv')
df2021=pd.read_csv('../data/interim/2021_consum_electricitat_bcn.csv')
df2022=pd.read_csv('../data/interim/2022_consum_electricitat_bcn.csv')
df2023=pd.read_csv('../data/interim/2023_consum_electricitat_bcn.csv')
df2024=pd.read_csv('../data/interim/2024_consum_electricitat_BCN.csv')

#Union de todos los DF
df = pd.concat([df2019, df2020, df2021, df2022, df2023, df2024], ignore_index=True)

df[df['Data']=='2019-01-20']
```

```python
df
```

```python
df.info()
```

```python
#Comprobación de que la columa data esta tratada en tipo datetime
df['Data'] = pd.to_datetime(df['Data'] , dayfirst=False)
#Transformar los datos valor a numerico
df['Valor'] = pd.to_numeric(df['Valor'], )
#Transformar a tipo str el sector
df['Sector_Economic'] = df['Sector_Economic'].astype(str)
#Eliminar valores no apreciables
df = df[df['Tram_Horari'] != 'No consta']
```

```python
#Crear un nuevo df con los valores necesarios
df_group = df.groupby(['Data', 'Sector_Economic'])['Valor'].sum().reset_index()
df_group.head(-5)
```

```python
#Eliminamos  el sector no especificat
df_group = df_group[df_group['Sector_Economic'] != 'No especificat']
df_group.shape
```

```python
df_group.head()
```

```python
cont_sector = df_group['Sector_Economic'].value_counts()
print(cont_sector)
```

```python
# Extraer el año de cada fecha
df_group['año'] = df_group['Data'].dt.year
# Contar cuántos días hay por año
cont_dias_año = df_group['año'].value_counts().sort_index()

print(cont_dias_año/3)
```

```python
dias_posteriores = df[df['Data'].dt.day > 12]
print("Datos de días después del 12:", dias_posteriores)
```
