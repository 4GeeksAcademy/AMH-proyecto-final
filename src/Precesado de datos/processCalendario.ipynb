## Este fichero nos permite procesar los datos del calendario de festivos creado en excel

```python
#Libraries
import pandas as pd
```

```python
#Cargar de los csv de cada estacion como df
df_festivos=pd.read_csv('../data/interim/base_calendario_final.csv')
df_COVID=pd.read_csv('../data/interim/COVID_DAY.csv', delimiter=';')

df_festivos
```

```python
df_COVID
```

```python
df_COVID = df_COVID.drop('Unnamed: 2',axis=1)
```

```python
df_festivos.info()
```

```python
df_COVID.info()
```

```python
#Transformamos la variable de tipo objeto a tipo fecha
df_festivos['fecha'] = pd.to_datetime(df_festivos['fecha'])
df_COVID['FECHA'] = pd.to_datetime(df_COVID['FECHA'], format='%d/%m/%Y')
```

```python
df_COVID
```

```python
# Unir los DataFrames, en funcion de la fecha
df_calendar = pd.merge(df_festivos, df_COVID, left_on='fecha', right_on='FECHA', how='inner')
```

```python
df_calendar=df_calendar.drop('FECHA', axis=1)
df_calendar
```

```python
#Guardamos los datos en un CSV
df_calendar.to_csv("../data/Final_Data/df_calendar.csv", index = False)
```
